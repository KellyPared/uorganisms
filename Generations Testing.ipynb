This is a simple genetics simulator in Python that will let students explore genetic reproduction over the course of many generations.

```python
import generations
from organisms import Organism
import pandas as pd
import matplotlib.pyplot as plt
```

This section defines some parameters of the simulation

* INITIAL_POP: The initial population size of organisms. Organisms in this generation have traits that are chosen at random
* GENERATIONS: The number of generations to simulate. Each organism will reproduce with another organism of the opposite gender in its generation
* POPULATION_CAP: the maximum number of organisms in a generation. Any extra organisms will be culled once the population cap is reached
* FAMILY_SIZE: the mean number of offspring from a reproductive pair
* FAMILY_STDEV: the standard deviation of family size.

```python
INITIAL_POP = 1000
GENERATIONS = 500
POPULATION_CAP = 2000
FAMILY_SIZE = 2.05
FAMILY_STDEV = 0.9
```

This section runs the simulation of each generation. The algorithm randomly selects a male and female organism for reproduction, resulting in a normalized number of offspring using the FAMLIY_SIZE and FAMILY_STDEV to define the distribution.

If there aren't an equal number of male and female organisms, or the total number of new organisms exceeds the population_cap, the remaining organisms are not allowed to reproduce.

```python
generations.test_run(INITIAL_POP, GENERATIONS, POPULATION_CAP, FAMILY_SIZE, FAMILY_STDEV)
```

# Trend Charts
First, we'll chart the total population size with a breakdown of the number of organisms with each color gene pair.

```python
overall_stats = generations.get_overall_stats()

# create a pandas dataframe from the overall statistics
df = pd.DataFrame.from_dict(overall_stats)
df.index = df.loc[:,'gen']

# plot the color genes
df.plot(y=['total_pop', 'BB', 'Bb', 'bb'])
plt.show()
```

```python
# plot the color trait percentages by generation
df['BluePct'] = df['blue'] / df['total_pop'] * 100
df['YellowPct'] = df['yellow'] / df['total_pop'] * 100

df.plot(y=['BluePct', 'YellowPct'])
plt.show()
```
